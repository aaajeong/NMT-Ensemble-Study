```python
wb = openpyxl.load_workbook('BLEU_Score.xlsx')
sheet = wb['Model 1 BLEU']
bleu1 = []
bleu2 = []
bleu3 = []
bleu4 = []

for i in range(2, 1002):
    bleu1.append(float(sheet.cell(row = i, column = 5).value))
    bleu2.append(float(sheet.cell(row = i, column = 6).value))
    bleu3.append(float(sheet.cell(row = i, column = 7).value))
    bleu4.append(float(sheet.cell(row = i, column = 8).value))
    
bleu1 = np.array(bleu1)
bleu2 = np.array(bleu2)
bleu3 = np.array(bleu3)
bleu4 = np.array(bleu4)

sheet.cell(row = 1002, column = 5).value = bleu1.mean()
sheet.cell(row = 1002, column = 6).value = bleu2.mean()
sheet.cell(row = 1002, column = 7).value = bleu3.mean()
sheet.cell(row = 1002, column = 8).value = bleu4.mean()

wb.save('BLEU_Score.xlsx')
```
